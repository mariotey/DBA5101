```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

NetReward = 0
NetPull = 0
number_machine = 24
```

```python
def pull(user_group, secret_key, machine_num):
    global NetReward
    global NetPull
    NetReward = NetReward + 1
    NetPull = NetPull + 1
    return {'Arm': str(machine_num), 'NetReward': NetReward, 'Pull' : NetPull, 'Reward':  np.random.randint(0, 100)}
```

```python
def epsilon_greedy (user_group, password, tot_rounds, number_machine, epsilon):    
    
    result = {idx + 1: {
        "Arm": idx + 1,
        "Output_Arm": 0,
        "Pulls": 0,
        "Reward_List": [],
        "Total_Reward": 0,
        "Average_Reward": 0
        } for idx in range(number_machine) }
    
    for num_round in range(1, tot_rounds + 1):        
        if random.random() < epsilon or num_round == 1:
            arm = random.randint(1, number_machine)
            print(f"# {num_round}, Exploring Arm {arm}...")
        else:
            arm = max(result, key=lambda k: sum(result[k]["Reward_List"]) / result[k]["Pulls"] if result[k]["Pulls"] > 0 else 0)
            print(f"# {num_round}, Exploiting Arm {arm}...")
            
        output = pull(user_group, password, arm)
        
        result[arm]["Output_Arm"] = output["Arm"]
        result[arm]["Reward_List"].append(output["Reward"])
        result[arm]["Pulls"] += 1
        result[arm]["Total_Reward"] = sum(result[arm]["Reward_List"])
        result[arm]["Average_Reward"] = 0 if result[arm]["Pulls"] == 0 else result[arm]["Total_Reward"]/result[arm]["Pulls"]
        
    return result
```

```python
result = epsilon_greedy("test_user","aaaaaaaa", 300, 24, 0.4)
if result:
    print("\nResults")
    for key in result.keys():
        print(result[key])
```
