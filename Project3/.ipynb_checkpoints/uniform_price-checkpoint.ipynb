```python
import numpy as np

NetReward = 515777
NetPull = 23729
tot_machine = 24
```

```python
def pull(user_group, secret_key, machine_num):
    global NetReward
    global NetPull
    
    NetReward = NetReward + 1
    NetPull = NetPull + 1
    
    return {'Arm': str(machine_num), 'NetReward': NetReward, 'Pull':NetPull, 'Reward':  np.random.randint(0, 100)}
```

```python
output = pull('test_user','aaaaaaaa',21)
output
```

```python
#################################################################################################

def uniform_price(user_group, secret_key, N, tot_rounds):
    
    # Initialize the result dict with arms and statistics
    result = {idx + 1: {
                        "Arm": idx + 1,
                        "Output_Arm": 0,
                        "Total": 0, 
                        "N_List": [], 
                        "Reward_List": [], 
                        "Total_Reward": 0,
                        # "Net_Reward": 0,
                        # "Net_Pull":0,
                        } for idx in range(tot_machine)
             }
    
    # Calculates the number of rounds left for Exploitation
    cut_off = tot_rounds - (tot_machine*N)
    
    # If number of rounds for Exploitation <= 0, do not pull
    if cut_off <= 0:
        print("Insufficient Total Rounds")
    
    else:
        arm = 0
        explore_flag = True
        
        for num_round in range(tot_rounds):
            # If iterating through Exploration rounds
            if (num_round + 1) <= (tot_machine*N):
                
                # Iterate through the machines 
                if arm == tot_machine:
                    print("\n")
                    arm = 1
                else:
                    arm = arm + 1
                
                explore_flag = True

            # If Exploration rounds are over, enter Exploitation rounds
            else:              
                print("\n")
                explore_flag = False
                
                # Identify the arm with the highest average Reward value across N_list
                arm = max(result, key=lambda k: sum(result[k]["N_List"]))
            
            # Simulate pulling an arm and getting an output
            output = pull(user_group,secret_key,arm)
            
            # Append Data
            if explore_flag == True:
                print(f"# {num_round + 1}, Exploring Arm {arm}...")
                result[arm]["N_List"].append(output["Reward"])
            else:
                print(f"# {num_round + 1}, Exploiting Arm {arm}...")
            
            result[arm]["Output_Arm"] = output["Arm"]
            result[arm]["Reward_List"].append(output["Reward"])
            result[arm]["Total"] = result[arm]["Total"] + 1
            # result[arm]["Net_Reward"] = output["NetReward"]
            # result[arm]["Net_Pull"] = output["Pull"]
        
        for key in result.keys():
            result[key]["Total_Reward"] = sum(result[key]["Reward_List"])
            
        return result

#################################################################################################
```

```python
result = uniform_price("test_user", "aaaaaaaa", 3, 100)

if result:
    print("\nResults")
    for key in result.keys():
        print(result[key])
```
