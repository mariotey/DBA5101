```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
ucb_df = pd.read_csv("UCB1_v2.csv")
uni_df = pd.read_csv("uniform_explor.csv")
greedy_df = pd.read_csv("epsilon_greedy_results.csv")
adapt_exp_df = pd.read_csv("adapt_exp.csv")
adapt_overt_df = pd.read_csv("adapt_overt.csv")
```

## All 1000 rounds

```python
plt.plot(uni_df["round"], uni_df["regret"], label="Uniform")
plt.plot(ucb_df["Unnamed: 0"], ucb_df["regret"], label="UCB")
plt.plot(greedy_df["Pull_Number"], greedy_df["Regret"], label="Greedy")
plt.plot(adapt_exp_df["round"], adapt_exp_df["regret"], label="Adapt(exp)")
plt.plot(adapt_overt_df["round"], adapt_overt_df["regret"], label="Adapt(1/t)")

plt.xlabel("No. of Rounds")
plt.ylabel("Total Regret")

plt.legend()

plt.show()
```

```python
plt.plot(uni_df["round"], uni_df["cum_reward"], label="Uniform")
plt.plot(ucb_df["Unnamed: 0"], ucb_df["reward"], label="UCB")
plt.plot(greedy_df["Pull_Number"], greedy_df["Cumulative_Reward"], label="Greedy")
plt.plot(adapt_exp_df["round"], adapt_exp_df["cum_reward"], label="Adapt(exp)")
plt.plot(adapt_overt_df["round"], adapt_overt_df["cum_reward"], label="Adapt(1/t)")

plt.xlabel("No. of Rounds")
plt.ylabel("Total Reward")

plt.legend()

plt.show()
```
